```python
#importing necessary matpotib libraries
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
#reading and showing the RGB image
image = mpimg.imread("owl.jpg")
imgplot = plt.imshow(image)
plt.show()
```

```python
#Conversion of RGB to Grayscale
R, G, B = image[:,:,0], image[:,:,1], image[:,:,2]
imgGray = 0.2989 * R + 0.5870 * G + 0.1140 * B
plt.imshow(imgGray, cmap='gray')
plt.show()
```

```python
#Conversion to HSV
hsv = matplotlib.colors.rgb_to_hsv(image)
plt.imshow(hsv)
```

```python
#importing OpenCV 
import numpy as np
import cv2
```

```python
image01 = cv2.imread("owl.jpg")
image01 = cv2.cvtColor(image01, cv2.COLOR_BGR2RGB)
grayscle_image = cv2.cvtColor(image01, cv2.COLOR_BGR2GRAY)
invertd_image = cv2.bitwise_not(grayscle_image)
smooth_image = cv2.GaussianBlur(invertd_image, (21, 21),sigmaX=0, sigmaY=0)
```

```python
pencil_sketch = cv2.divide(grayscle_image, 255 - smooth_image, scale=255)
plt.imshow(pencil_sketch, cmap="gray")
plt.title("Pencil Sketch Image")
plt.show()
```
